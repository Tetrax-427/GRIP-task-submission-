From given "IRIS" dataset ,pridicting optimal number of clusters and representing it grahycally.

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
 # loading dataset
```

```python
X = iris.data[:, :2]         
y = iris.target         

# breaking dataset in X, Y for traning
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel('Sepal Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)

#scatter plot  Sepal length V/s Sepal Width
```

Finding Optimal value of K(number of clusters)
using Knee Method

```python
cost =[]
for i in range(1, 11):
    KM = KMeans(n_clusters = i, max_iter = 500)
    KM.fit(X)
      
    # calculates squared error
    # for the clustered points
    cost.append(KM.inertia_)    
    
  
plt.plot(range(1, 11), cost, color ='r', linewidth ='3')
plt.xlabel("Value of K")
plt.ylabel("Sqaured Error (Cost)")
plt.show()
```

Using "3" as value for number of clusters

```python
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```
